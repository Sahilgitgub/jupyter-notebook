```python
# Data cleaning means 
# fixing bad data in your data set.
# Bad data could be:
# Empty cells
# Data in wrong format
# Wrong data
# Duplicates
```

```python
# cleaning empty cells
df=pd.read_csv('data.csv')
# new_df=df.head(25)
new_df=df.dropna() #dropna() method returns a new DataFrame, and will not change the original and removes the nan values from dataset.
print(new_df.to_string())
```

```python
# jaha jaha calories m nan values h wha 130 fill kardega
df=pd.read_csv('data.csv')
df['Calories'].fillna(130,inplace=True)
print(df.to_string())
```

```python
df=pd.read_csv('data.csv')
df['Maxpulse'].fillna(20,inplace=True)
print(df.to_string())
```

```python
# Calculate the MODE, mean ,median and replace any empty values with it.
df=pd.read_csv('data.csv')
x=df['Calories'].mean()
df['Calories'].fillna(x,inplace=True)
print(df.to_string())
```

```python
y=df['Calories'].median()
df['Maxpulse'].fillna(y,inplace=True)
print(df.to_string())
```

```python
z=df['Pulse'].mode()
df['Pulse'].fillna(z)
print(df.to_string())
```
